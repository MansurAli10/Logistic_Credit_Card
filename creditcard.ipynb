```python
import pandas as pd
```

```python
cc_df=pd.read_csv('creditcard.csv')

cc_df
```

```python
cc_df.shape
```

# Train test split

```python
cc_df.columns
```

```python
y=cc_df['Class']
y
```

```python
cc_df.drop(columns=['Class'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(cc_df,y, test_size=0.2,random_state=77)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression()
```

```python
log_reg.fit(X_train,y_train)
```

```python
pred=log_reg.predict(X_test)

pred
```

```python
train_pred=log_reg.predict(X_train)

train_pred
```

```python
probabs=log_reg.predict_proba(X_test)

probabs
```

```python
pos_probabs=probabs[::,1]

pos_probabs
```

```python
from sklearn.metrics import roc_curve
```

```python
fpr,tpr,thresholds=roc_curve(y_test,pos_probabs)

fpr,tpr,thresholds
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(fpr,tpr)
```
